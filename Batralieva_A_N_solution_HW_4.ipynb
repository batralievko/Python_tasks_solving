````python
# Ваше решение здесь

import requests
from bs4 import BeautifulSoup
import re
url = 'https://stepik.org/media/attachments/lesson/258943/Moscow.html'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'lxml')
text = soup.get_text()
refs = []
for i in soup.find_all('a'):
    a = i.get('href')
    if a == None:
        continue
    else:
        if ':' in a:
            continue
        else:
            if '#' in a:
                continue
            else:
                if 'http://' in a:
                    continue
                else:
                    if 'https://' in a:
                        continue
                    else:
                        refs.append(a)
print('Число внутренних ссылок:', len(refs))
````

### Задание 4. Скрейпинг таблицы (2 балла)

В файле https://stepik.org/media/attachments/lesson/209723/3.html находится одна таблица. Просуммируйте все числа в ней и введите в качестве ответа одно число - эту сумму. Для доступа к ячейкам используйте возможности BeautifulSoup.


````python
import requests
from bs4 import BeautifulSoup
import re
page = requests.get('https://stepik.org/media/attachments/lesson/209723/3.html')
soup = BeautifulSoup(page.text, 'lxml')
text = soup.get_text()
numbers = []
for i in soup.find_all('tr'):
    a = i.find_all('td')
    for j in a:
        num = int(j.text.strip())
        numbers.append(num)
print(sum(numbers))
````
